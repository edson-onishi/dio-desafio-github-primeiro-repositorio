# **Tuplas**


```python
#As tuplas usam parenteses como sintaxe
tp = ("Banana","Maçã",10, 50)
```

```python
#Retornando o primeiro elemento
tp[0]
```

```python
#DIferente das istas as tuplas são imutáveis, o que quer dizer que não podemos alterar os seus elementos
tp[0] = "Laranja"
```

```python
tp.count("Maçã")
```

```python
tp[0:2]
```

#**Dicionários**

```python
#Para criar um dicionário utilizamos as {}
dc = {"Maçã":20, "Banana":10, "Laranja":15, "Uva":5}#Dicionários trabalham com o conceito chave e valor
```

```python
dc
```

```python
#Acessando o valor de um dicionário através da chave
dc["Maçã"]
```

```python
#Atualizando o valor da Maçã
dc["Maçã"] = 25
dc
```

```python
#Reornando todas as chaves do dicionário
dc.keys()
```

```python
#Retornando os valores do dicionarios
dc.values()
```

```python
#Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("Limão", 22)
```
